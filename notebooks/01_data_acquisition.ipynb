Scrape Billboard Hot 100 labels and fetch Spotify 30s preview URLs.

```python
import requests
import pandas as pd
import time
from dotenv import load_dotenv
import os
```

```python
load_dotenv("../.env")

CLIENT_ID = os.getenv("SPOTIFY_CLIENT_ID")
CLIENT_SECRET = os.getenv("SPOTIFY_CLIENT_SECRET")

# TODO: authenticate with Spotify client credentials flow
token = None
```

```python
df = pd.read_csv("../data/raw/hot100.csv")
df = df.rename(columns={
    "Song": "title",
    "Artist": "artist",
    "Peak Position": "peak_pos",
    "Weeks in Charts": "weeks_on_chart"
})
df["label"] = (df["peak_pos"] <= 10).astype(int)
df = df[["title", "artist", "peak_pos", "weeks_on_chart", "label"]].drop_duplicates(subset=["title", "artist"])
df.head()
```
