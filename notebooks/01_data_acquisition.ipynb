# 01 — Data Acquisition

```python
import polars as pl
```

```python
df = pl.read_csv("../data/raw/hot100.csv")
df = df.rename({
    "Song": "title",
    "Artist": "artist",
    "Peak Position": "peak_pos",
    "Weeks in Charts": "weeks_on_chart"
})
df = df.with_columns((pl.col("peak_pos") <= 10).cast(pl.Int8).alias("label"))
df = df.select(["title", "artist", "peak_pos", "weeks_on_chart", "label"]).unique(subset=["title", "artist"])
df.head()
```

```python
print(df["label"].value_counts())
print(f"Hit ratio: {df['label'].mean():.1%}")
```
